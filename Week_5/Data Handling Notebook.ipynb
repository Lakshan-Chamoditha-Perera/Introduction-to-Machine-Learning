```python
df_titanic['Fare'] = df_titanic.Fare.str.replace('$', ' ').astype(float)
```

```python
df_titanic['Sex_endswith'] = list(map(lambda x: x.endswith(' '), df_titanic['Sex']))
df_titanic
```

```python
df_titanic['Sex_startswith'] = list(map(lambda x: x.startswith(' '), df_titanic['Sex']))
df_titanic
```

```python
print(df_titanic[df_titanic['Sex_endswith'] == True])
```

```python
df_titanic.drop(['Sex_endswith', 'Sex_startswith'], axis=1, inplace=True)
```

```python
df_titanic['Sex_new'] = list(map(lambda x: x.strip(), df_titanic['Sex']))
df_titanic
```

```python
# plot
import seaborn as sns 
sns.countplot(x='Sex_new', data=df_titanic)
```

```python
df_titanic.Sex_new.value_counts()
```

```python
# map each value to either "male or female"
df_titanic['Sex_new'] = (df_titanic['Sex_new']
                         .map({'male': 'male', 'female': 'female', 'fem': 'female', 'mal': 'male'}))
```
