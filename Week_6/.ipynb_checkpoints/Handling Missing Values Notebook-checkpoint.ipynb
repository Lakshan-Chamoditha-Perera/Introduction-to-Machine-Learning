```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("data/04 Handling Missing Data using Python_Saless.xlsx")
# check missing values
df.isnull().sum()
```

```python
# load titanic data 
df_titanic = pd.read_csv("data/03 Handling Structural Error_titanic_train.csv")
df_titanic.describe() # check missing values
```

```python
import missingno as msno
msno.matrix(df_titanic)
```

```python
msno.bar(df_titanic,figsize=(10,4),fontsize=12)
```

```python
plt.figure()
sns.heatmap(df_titanic.isnull(),cmap=sns.color_palette(['black','yellow','orange','white']),cbar=False)
```
